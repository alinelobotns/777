# **Dicionário de Variáveis**


As variáveis de estudo do DataFrame serão;


`Álbuns` - Todos os álbuns de estúdio, Remix, Acústicos, Compilações e Extended plays,

`Músicas` - Todas as músicas e parcerias,

`Streams` - Os números de reproduções com 30 segundos ou mais, na plataforma Spotify,

`Crescimento`- Crescimento de reproduções, referente ao dia 08/02/2023.


```python
#Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import seaborn as sns

sns.set()

df = pd.read_excel('/content/Justin Bieber Spotify musicas.xlsx')
df2 = pd.read_excel('/content/Justin Bieber Spotify.xlsx')


```

Abaixo, podemos observar um gráfico de barras com as 20 músicas mais reproduzidas do catálogo do Spotify


```python
top20 = df

sns.set_style('darkgrid')
plt.figure(figsize=(15,5))
plt.title('Top 10')
plt.xticks(rotation=90)


sns.barplot(data = top20.sort_values('Streams', ascending=False).head(20), x = 'Músicas', y='Streams')

;


```

Pelo método `max`, podemos observar que a música **Stay With Justin Bieber**  tem 23.84.580.729 de reproduções, até o momento.

```python
df.Streams.max()
```

No gráfico de linha, podemos notar que a música *Cold Water*, lançada no dia 22 de julho de 2016, teve um crescimento considerável, no dia 8 de fevereiro de 20233 

```python
sns.set_style("whitegrid")
plt.figure(figsize=(15,5))
plt.title('Crescimento dia 08/02/2023')
plt.xticks(rotation=90)

sns.lineplot(data = df.sort_values('Crescimento', ascending=False).head(20), x = 'Músicas', y='Crescimento', color= 'G');

plt.show();
```

Pelo método `max`, é informado que em 08 de fevereiro de 2023, o Justin Bieber, teve 1.433.151.976 de reprodução a mais

Pelo gráfico WordCloud, podemos identificar as palavras mais mencionadas em nosso Dataset 

```python
#importando a biblioteca de wordcloud

!pip install wordcloud -q
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# A quantidade de valores ausentes
print("Valores ausentes para Música: ", df.Músicas.isnull().sum())

# eliminar as colunas com valores ausentes
Músicas = df.dropna(subset=['Álbuns'], axis=0)['Músicas']
```

```python
all_Músicas = " ".join(s for s in Músicas)

print("Quantidade de Palavras: {}".format(len(all_Músicas)))
```

```python
# endereço da imagem
JB = np.array(Image.open("/content/justin_bieber photoshop ele preto.jpg"))

stopwords = set(STOPWORDS)
stopwords.update(["feat", "Remix", "Acoustic", "live", "Version", "Instrumental", "Bonus", 
                  "PAULO", "JACKINSKY", "Joe", "Gauthreaux", "Chandler" "Moore", "Radio", "Mix" "bot", 
                  "audiobot", "Club Mix", "Club", "vice", "Peter", "SuperFestive", "Barona", " DJ Laszlo"])

# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1000, height=1000, max_words=2000,
                      mask=JB, max_font_size=400,
                      min_font_size=4).generate(all_Músicas)
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(20,20))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
plt.imshow(wordcloud)

wordcloud.to_file("JB.png");
```
